```python
# 導入套件
import requests as req
import json
import re
from bs4 import BeautifulSoup
from time import sleep
from random import ra
```

```python
# 要爬的網址
url = 'https://www.gutenberg.org/browse/languages/zh'
# 模擬瀏覽器的header
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}
# 正規表達式中文以及標點符號(部份而已，有點懶得全打)
regex = r"[\u4E00-\u9FFF]*[\u3002]?[\uff0c]?[\uff1f]?[\uff01]?[\uff1a]?[\u300c]?[\u300d]?[\uff1b]?[\u3001]?"
# 接收request回來的東西 有收到就是200
res = req.get(url)
# 美麗湯解析
soup = BeautifulSoup(res.text, 'lxml')
# 使用CSS選擇器抓所有的書名
book_list = soup.select('div.pgdbbylanguage li')
for book in book_list:
    # 確定是中文書
    if book.get_text()[-20:-13] == 'Chinese':
        bookName = book.select_one('a').get_text()
        # 取得書籍網址編號
        bootUrlNum = book.select_one('a')['href'].split('/')[2]
        # 文章內容的連結
        bootUrl = f'https://www.gutenberg.org/files/{bootUrlNum}/{bootUrlNum}-0.txt'
        book_res = req.get(bootUrl)
        # 文章內容編碼
        book_res.encoding='utf8'
        book_soup = BeautifulSoup(book_res.text, 'lxml')
        book_content = book_soup.select('body')
        # 檔案寫入
        fileout1 = open(f'rawdata/{bookName}.txt','w')
        for c in book_content:
            fileout1.writelines(re.findall(regex,c.get_text()))
        fileout1.close()
        print(f'Download {bookName}{bootUrl} sueecss')
```
