```python
import requests as req
import json, re
from bs4 import BeautifulSoup as bs
from time import sleep
from random import randint
```

```python
regex_month = r'[01]?[0-9]{1}\s\S'

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}

url = 'https://streetvoice.com/venue/activities/all/0/'
res = req.get(url, headers = headers)
soup = bs(res.text, 'lxml')
item_list = soup.select('div#item_box_list > div')
for item in item_list:
    item_month = re.findall(regex_month, item.select_one('div.bg-red.text-white.pt-1.pb-1').get_text())[0].replace(' ', '')
    item_days = item.select_one('h1').get_text()
    item_dow = item.select_one('h4').get_text()
    print(f'{item_month}{item_days}號{item_dow}')
    events = item.select('ul.list-group.list-group-flush.list-group-live > li')
    for event in events:
        print(event.select('h3')[0].get_text())
        print(event.select('h4')[0].get_text())
        place = event.select('h4')[0].get_text()[6:]
        g_url = f"https://www.google.com.tw/search?tbm=map&authuser=0&hl=zh-TW&gl=tw&q={place}"
        g_res = req.get(g_url, headers = headers)
        data = json.loads(g_res.text[5:])
        print(data[1][0][2], data[1][0][1])
        sleep(randint(1,2))
    print('-' * 50)
        
```
