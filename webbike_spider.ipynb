```python
import requests as req
import json
from bs4 import BeautifulSoup as bs
from time import sleep
from random import randint
```

```python
search_input = '安全帽'
url = f"https://www.webike.tw/parts?q={search_input}"

headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'
}

res = req.get(url)
soup = bs(res.text, 'lxml')
pages = soup.select('div.d-flex.align-items-center.justify-content-around > span')[0].get_text()
#soup.select('div.col-6.col-md-4.col-lg-4.col-xl-3.px-2.mb-3 h4.card-title.h6.hide-line-2.item__body__title.text-left.mb-1.js-product-name')[0]['title']
pages = int(pages.split(' ')[3])
soup
```

```python
for page in range(1, pages):
    url = f"https://www.webike.tw/parts?q={search_input}&page={page}"
    res = req.get(url)
    soup = bs(res.text, 'lxml')
    items = soup.select('div.col-6.col-md-4.col-lg-4.col-xl-3.px-2.mb-3')
    for item in items:
        print(f"product_name: {item.select_one('h4.card-title.h6.hide-line-2.item__body__title.text-left.mb-1.js-product-name')['title']}")
        print(f"price: {item.select_one('strong')['title']}")              
        print('-' * 50)
    sleep(randint(1,3))
```
