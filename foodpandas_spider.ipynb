```python
import requests as req
import json
```

```python
my_place = '資展國際'
url = f'https://www.google.com.tw/search?tbm=map&authuser=0&hl=zh-TW&gl=tw&q={my_place}'
res = req.get(url)
data = json.loads(res.text[5:])
lon = data[1][0][1]
lat = data[1][0][2]
print(lon, lat)
```

```python
fp_url = 'https://disco.deliveryhero.io/listing/api/v1/pandora/vendors'
query = {
    'longitude': lon,  # 經度
    'latitude': lat,  # 緯度
    'language_id': 6,
    'include': 'characteristics',
    'dynamic_pricing': 0,
    'configuration': 'Variant1',
    'country': 'tw',
    'budgets': '',
    'cuisine': '',
    'sort': 'distance_desc',
    'food_characteristic': '',
    'use_free_delivery_label': False,
    'vertical': 'restaurants',
    'opening_type': 'pickup',
    'limit': 200,
    'offset': 0,
    'customer_type': 'regular'
}
headers = {
    'x-disco-client-id': 'web',
}
res = req.get(url=fp_url, params=query, headers=headers)
data = json.loads(res.text)
restaurants = data['data']['items']
for restaurant in restaurants:
    print(f"name: {restaurant['name']}, addr: {restaurant['address']} latitude:{restaurant['latitude']}, longitude:{restaurant['longitude']}")
    print(f"{restaurant['distance']*1000}m")
    print('-'*50)
```
